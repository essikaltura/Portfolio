```python
#Nimeän sarakkeet uudelleen lyhyemmillä nimillä
df2 = df2.rename(columns={'Espoon portti (Eco-Counter)':'Espoon portti','Espoonlahdenraitti (Eco-Counter)':'Espoonlahdenraitti',
                    'Gallen-Kallela (Eco-Counter)':'Gallen-Kallela'})
```

```python
#Laitan päivämäärät muotoon, jonka Pyton ymmärtää
df2.index = pd.to_datetime(df2['Päivämäärä'])
```
